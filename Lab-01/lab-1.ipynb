```python
# Task 01 - 2

import numpy as np
import matplotlib.pyplot as plt

# Generate random arrays from different distributions
uniform_data = np.random.rand(100000, 1)
normal_data = np.random.randn(100000, 1)
exponential_data = np.random.exponential(scale=1, size=(100000, 1))

# Plot histograms
plt.hist(uniform_data, density=True, bins=50, histtype="step", color="blue", label="Uniform")
plt.hist(normal_data, density=True, bins=50, histtype="step", color="orange", label="Normal")
plt.hist(exponential_data, density=True, bins=50, histtype="step", color="green", label="Exponential")
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.xlim([-2.5, 10])  # Adjust x-axis range for better visualization
plt.ylim([0, 1.1])
plt.show()
```

```python
# Task 01 - 3

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a grid of x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calculate the corresponding Z values
Z = X**2 + Y**2

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surface = ax.plot_surface(X, Y, Z, cmap='viridis')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Z = X^2 + Y^2')

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset (replace 'your_dataset.csv' with the actual file name) 
df = pd.read_csv('C:/Users/JanithGamage/Desktop/Y4S2/DL/Lab/Week-1/Pokemon.csv', index_col=0, encoding='latin')

# For demonstration, let's create a sample DataFrame
data = {
    'HP': np.random.rand(100),
    'Attack': np.random.rand(100),
    'Defense': np.random.rand(100),
    'Sp. Atk': np.random.rand(100),
    'Sp. Def': np.random.rand(100),
    'Speed': np.random.rand(100)
}
df = pd.DataFrame(data)

# Calculate Pearson correlation matrix
pearson_corr = df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].corr(method='pearson')

# Calculate Spearman rank correlation matrix
spearman_corr = df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].corr(method='spearman')

# Create subplots for heatmaps
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot Pearson correlation heatmap
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', center=0, ax=ax1)
ax1.set_title('Pearson Correlation')

# Plot Spearman correlation heatmap
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', center=0, ax=ax2)
ax2.set_title('Spearman Correlation')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
 
# Load the dataset (replace '.csv' with the actual file name) 
df = pd.read_csv('C:/Users/JanithGamage/Desktop/Y4S2/DL/Lab/Week-1/Au_nanoparticle_dataset.csv', index_col=0, encoding='latin')


# 1. Create a new dataframe with specified columns
new_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]

# 2. Display the first 20 samples
print(new_df.head(20))

# 3. Calculate mean, standard deviation, and quartile values
summary_stats = new_df.describe()
print(summary_stats)

# 4. Plot histograms of each feature in a 1x4 layout
plt.figure(figsize=(12, 3))
for i, col in enumerate(new_df.columns):
    plt.subplot(1, 4, i+1)
    plt.hist(new_df[col], bins=20, edgecolor='black')
    plt.title(col)
plt.tight_layout()
plt.show()

# 5. Visualize scatter plots and histograms using pairplot
sns.pairplot(new_df)
plt.show()

# 6. Customized pairplot with kdeplot and histograms
g = sns.PairGrid(new_df)
g.map_upper(sns.histplot)  # Bivariate histogram
g.map_diag(sns.histplot)
g.map_lower(sns.kdeplot)
plt.show()
```
